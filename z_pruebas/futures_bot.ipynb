```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
eth_1d = pd.read_csv(r'/Users/fernandoseguias/Desktop/ferseg/Proyectos/Programacion/iron_hack/project_1/output/eth_1d.csv')
```

```python
eth_1d["increase_decrease"] = (eth_1d.close_price - eth_1d.open_price) / eth_1d.open_price * 100
eth_1d.tail(5)
```

```python
eth_1d_180 = eth_1d.tail(180)
```

```python
green_candles = []
red_candles = []

def split_candles(x):
    if x > 0:
        green_candles.append(x)
    elif x < 0:
        red_candles.append(x)
    else:
        pass

eth_1d_180["increase_decrease"].apply(split_candles)
box_plot_data = [pd.DataFrame(green_candles)[0], pd.DataFrame(red_candles)[0].apply(lambda x : -x)]
plt.boxplot(box_plot_data)
plt.show()

```
