```python
import requests
import pandas as pd
import numpy as np
from time import sleep
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import warnings
warnings.filterwarnings('ignore')

#------------------------------------------------

opciones= Options()
opciones.add_experimental_option('excludeSwitches', ['enable-automation']) #para ocultarme como robot
opciones.add_experimental_option('useAutomationExtension', False)
#opciones.add_argument('--start-maximized') #empezar maximizado
opciones.add_argument('user.data-dir=selenium') #guarda las cookies
opciones.add_argument('--incognito')#incognito window

#------------------------------------------------
```

```python
#create lists to store scrapped data
id_list = []

#get url list to scrap
url_list = []
for page in range(100):
    if page == 0:
        url = "https://app.traitsniper.com/snoop-dogg-doggies?limit=100"
        url_list.append(url)
    else:
        url = f"https://app.traitsniper.com/snoop-dogg-doggies?limit=100&page={page}"
        url_list.append(url)

#open url's - rarity check to scrap

for i in url_list[:10]:
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get(i)
    sleep(15)
    items_list = driver.find_elements_by_class_name('id_id__1y_VJ')
    sleep(5)
    for i in items_list:
        id_list.append(i.text)
    sleep(3)
    driver.quit()
id_list
```
